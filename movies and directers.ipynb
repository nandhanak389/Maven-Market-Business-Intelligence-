```python
import numpy as np
```

```python
import pandas as pd
directors=pd.read_csv("directors.csv")
directors
```

```python
movies=pd.read_csv("movies.csv")
movies
```

```python
a=pd.merge(movies,directors,left_on="director_id",right_on="id",how="left")  #left_on & right _on are used to where the column name are different cloumnnmaes and contain same datas 
a
```

```python
a.drop(columns=["Unnamed: 0_x"],inplace=True)
a
```

```python
a.drop(columns=["Unnamed: 0_y"],inplace=True)
a
```

```python
a.drop(columns=["id_y"],inplace=True)
a
```

```python
a.rename(columns={"id_x":"movie_id"},inplace=True)
a
```

```python
a.isna().sum()

```

```python
a[a["gender"].isnull()]
```

```python
a["gender"]=a["gender"].fillna("Male")
```

```python
a
```

```python
a.isna().sum()
```

### BASED ON MOVIES VOTE

```python
import matplotlib.pyplot as plt
```

```python
b=a.groupby("title")["vote_count"].max().sort_values(ascending=False).head(10).reset_index()
plt.figure(figsize=(10,5))
plt.xlabel("vote counts",color="orange",fontsize=12)
plt.ylabel("movies",color="orange",fontsize=12)
plt.title("movies based on vote",color="blue",fontsize=16,fontweight="bold",fontstyle="italic")
plt.plot(b["vote_count"],b["title"],marker="o",color="pink",linestyle="--",markerfacecolor="g");
```

### BASED ON MOVIES REVENUE

* In the comparison of BASED ON MOVIES VOTE and BASED ON MOVIES REVENUE we can see the differ of most rated movie and most revenue geted movies are not same
* Here the most rated movies can have best story 
* And the most revenue get movies is its making and graphics and vfx of the movie is great
* The revenue based movie = Avatar
* The vote count movie = 	Inception

```python
revenue=a.groupby("title")["revenue"].max().sort_values(ascending=False).head(10).reset_index()
plt.figure(figsize=(12,6))
plt.xlabel("revenue",color="orange",fontsize=12)
plt.ylabel("movies",color="orange",fontsize=12)
plt.title("movies based on revenue",color="blue",fontsize=16,fontweight="bold",fontstyle="italic")
plt.plot(revenue["revenue"],revenue["title"],marker="*",color="purple",linestyle="--",markerfacecolor="red")
plt.show()
```

### violin plot
* 

```python
vote_male = a[a['gender'] == 'Male']['vote_average']
vote_female = a[a['gender'] == 'Female']['vote_average']


plt.figure(figsize=(13,6))
plt.grid(True)
plt.xticks([1, 2], ['Male', 'Female'])
plt.violinplot([vote_male, vote_female])
plt.xlabel('Director Gender', color='tomato', fontsize=15, fontweight='bold', fontstyle='italic')
plt.ylabel('Vote Average', color='tomato', fontsize=15, fontweight='bold', fontstyle='italic')
plt.title('Distribution of Vote Average by Director Gender', color='orangered', fontsize=20, fontweight='bold', fontstyle='italic', fontname="Courier New")
plt.show()
```

### box plot
* max of the male profit distribution is high
* female is low

```python
budgets_male = a[a['gender'] == 'Male']['profit']
budgets_female = a[a['gender'] == 'Female']['profit']

plt.figure(figsize=(12,6))
plt.boxplot([budgets_male, budgets_female], tick_labels=['Male', 'Female'])
plt.grid(True)
plt.xlabel('Director_Gender', fontsize=15, fontweight='bold', color='darkgreen')
plt.ylabel('profit', fontsize=15, fontweight='bold', color='darkgreen')
plt.title('profit Distribution by Director Gender', fontsize=20, fontweight='bold', color='purple')
plt.show()
```
